```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("../../data/diabetes.csv")
```

```python
df.head()
```

```python
df_selected = df[[ 'Glucose', 'BloodPressure',
       'BMI','DiabetesPedigreeFunction', 'Age','Outcome']]
```

```python
features = ['Glucose', 'BloodPressure', 'BMI', 'DiabetesPedigreeFunction', 'Age']

for col in features:
    sns.histplot(df_selected[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
sns.countplot(x='Outcome', data=df_selected)
plt.title("Diabetes Distribution")
plt.show()
```

```python
for col in features:
    sns.boxplot(x='Outcome', y=col, data=df_selected)
    plt.title(f"{col} vs Outcome")
    plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_selected.corr(), annot=True, cmap='coolwarm')
plt.show()
```
